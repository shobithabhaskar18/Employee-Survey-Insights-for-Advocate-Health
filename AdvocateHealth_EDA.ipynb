```python
import matplotlib.pyplot as plt
import seaborn as sns

# Analyze trends in "Reasons to Leave" by Tenure, Market, and Region
# Count occurrences of key dissatisfaction factors in comments

# Define key words related to dissatisfaction
keywords = {
    "workload": ["workload", "burnout", "stressful", "overwhelming"],
    "compensation": ["compensation", "salary", "benefits", "underpaid"],
    "leadership": ["management", "leadership", "supervisor", "policies"],
    "career_growth": ["career", "growth", "opportunities", "limited"],
}

# Function to count occurrences of keywords in a column
def count_keyword_occurrences(column, keyword_list):
    return column.str.contains('|'.join(keyword_list), case=False, na=False).astype(int)

# Apply keyword count to "comment_reason_to_leave" column
for category, words in keywords.items():
    df[category] = count_keyword_occurrences(df["comment_reason_to_leave"], words)

# Group by tenure bands, market, and region
tenure_trends = df.groupby("Tenure Bands")[list(keywords.keys())].sum()
market_trends = df.groupby("Market")[list(keywords.keys())].sum()
region_trends = df.groupby("Region")[list(keywords.keys())].sum()

# Plot trends by tenure bands
plt.figure(figsize=(10, 5))
tenure_trends.plot(kind="bar", stacked=True, figsize=(10, 6))
plt.title("Reasons for Leaving by Tenure Bands")
plt.xlabel("Tenure Bands")
plt.ylabel("Count of Mentions")
plt.legend(title="Dissatisfaction Factors")
plt.xticks(rotation=0)
plt.show()

# Plot trends by market
plt.figure(figsize=(10, 5))
market_trends.plot(kind="bar", stacked=True, figsize=(10, 6))
plt.title("Reasons for Leaving by Market")
plt.xlabel("Market")
plt.ylabel("Count of Mentions")
plt.legend(title="Dissatisfaction Factors")
plt.xticks(rotation=45)
plt.show()

# Plot trends by region
plt.figure(figsize=(10, 5))
region_trends.plot(kind="bar", stacked=True, figsize=(10, 6))
plt.title("Reasons for Leaving by Region")
plt.xlabel("Region")
plt.ylabel("Count of Mentions")
plt.legend(title="Dissatisfaction Factors")
plt.xticks(rotation=0)
plt.show()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Select the relevant text columns for sentiment clustering
text_columns = [
    "comment_reason_to_stay",
    "comment_reason_to_leave",
    "comment_well_being_at_work",
    "comment_well_being_outside_work",
    "comment_burnout_reason",
    "comment_burnout_improvement",
    "comment_what_is_important_for_us_to_know"
]

# Function to clean text data
def preprocess_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W', ' ', text)  # Remove non-word characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    tokens = word_tokenize(text)  # Tokenization
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
    return " ".join(tokens)

# Apply preprocessing to all text columns
for col in text_columns:
    df[col] = df[col].apply(preprocess_text)

# Combine all text columns into a single column for clustering
df["combined_text"] = df[text_columns].apply(lambda x: ' '.join(x), axis=1)

# Display preprocessed text samples
df[["combined_text"]].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Alternative text preprocessing function without NLTK dependencies
def preprocess_text_alternative(text):
    if pd.isna(text):
        return ""
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W', ' ', text)  # Remove non-word characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Apply preprocessing to all text columns
for col in text_columns:
    df[col] = df[col].apply(preprocess_text_alternative)

# Combine all text columns into a single column for clustering
df["combined_text"] = df[text_columns].apply(lambda x: ' '.join(x), axis=1)

# Convert text data to numerical representation using TF-IDF
vectorizer = TfidfVectorizer(max_features=500, stop_words='english')
X = vectorizer.fit_transform(df["combined_text"])

# Check the shape of the transformed text data
X.shape
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the Elbow method
inertia = []
k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Choosing the optimal number of clusters (based on the Elbow Method, assuming k=4 as an example)
optimal_k = 4

# Applying K-Means clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X)
import pandas as pd

# Selecting relevant columns to display
clustered_data = df[["combined_text", "cluster"]]

# Show the first few rows of clustered data
print(clustered_data.head(20))  # Prints the first 20 rows
```
